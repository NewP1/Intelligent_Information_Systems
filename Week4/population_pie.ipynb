```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
plt.pie(size, labels = label, autopct = '%.1f%%')
plt.legend()
plt.show()
```

```python
plt.axis('equal')
size = [120, 250, 330, 410]
label=['레이블1', '레이블2', '레이블3', '레이블4']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
exp = (0,0,0.1,0)
plt.pie(size, labels = label, autopct = '%.2f%%', colors=color, explode=exp)
plt.legend()
plt.show()
```

```python
import csv

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

result = []
m_sum, f_sum = 0, 0

for row in data:
    if '송도1동' in row[0]:
        for i in range(101):
            m_sum = m_sum + int(row[i+3])
            f_sum += int(row[i+106])
            
result.append(m_sum)
result.append(f_sum)
print(result)
```

```python
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')

label = ['남', '여']
color = ['crimson', 'darkcyan']
plt.pie(result, labels=label, autopct='%.1f%%', colors=color, startangle=90)
plt.legend()
plt.show()
```
