```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male = []
female = []
result = []

for row in data:
    if '송도1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            result.append(int(m) + int(fm))
            male.append(-(int(m)))
            female.append(int(fm))
f.close()

plt.style.use('ggplot') # 격자 무늬 스타일 지정
plt.rc('font', family='Malgun Gothic')
plt.title('송도1동 연령별 인구수')


plt.bar(range(101), result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
plt.title('송도1동 연령별 인구수')

plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), male, label ='남성')
plt.barh(range(101), female, label='여성')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()
plt.show()
```
