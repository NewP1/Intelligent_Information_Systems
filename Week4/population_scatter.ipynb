```python
import random

x, y = [], []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
    
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=.3)
plt.show()
```

```python
import csv
import math

f = open('incheon_population.csv', 'r')
data = csv.reader(f)

male, female = [], []
size = []

for row in data:
    if '송도1동' in row[0]:
        for i in range(3, 104):
            m, fm = int(row[i]), int(row[i+103])
            male.append(int(row[i]))
            female.append(int(row[i+103]))
            size.append(math.sqrt(m+fm))
        break

```

```python


plt.rc('font', family = 'Malgun Gothic')
plt.title('연령대별 성별 인구 그래프')
plt.scatter(male, female, s=size, c=range(101), alpha=.9, cmap='jet')
plt.colorbar()


max_val = max(max(male), max(female))
plt.axis([0, max_val, 0, max_val])
line = range(max_val)

plt.plot(line, line, color='g')

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
