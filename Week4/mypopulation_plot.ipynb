```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)


result = []

for row in data:
    if '동춘1동' in row[0]:
        for m, fm in zip(row[3:104], row[106:207]):
            result.append(int(m) + int(fm))
f.close()


plt.rc('font', family='Malgun Gothic')
plt.title('동춘1동 연령별 인구수')


plt.style.use('ggplot') # 격자 무늬 스타일 지정
plt.xlabel('연령')
plt.ylabel('인구수')
plt.plot(result)

plt.show()
```

내가 사는 동춘1동은 50 ~ 60대 사람들이 가장 많고 그 다음으로는 40대 사람들이 많다.

20 ~ 30대는 점차 증가하는 추세를 보이고 70대 이후부터는 사람들이 점차 줄어드는 것을 볼 수 있다.
