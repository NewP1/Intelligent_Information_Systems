```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
next(data)
high = []                             # 최고 기온 데이터를 저장할 빈 리스트
low = []                              # 최저 기온 데이터를 저장할 빈 리스트

for row in data:                       
    if row[-1] != '' and row[-2] != '':                  # 최고 기온과 최저 기온 데이터 값이 존재한다면
        if int(row[0].split('-')[0] >= '1999'):           # 출생 년도 이후 데이터 비교
            if row[0].split('-')[1] == '07' and row[0].split('-')[2] == '12':   # 9월 13일
                high.append(float(row[-1]))   # result 리스트에 최고 기온 값 추가
                low.append(float(row[-2]))
        
f.close()
```

## 그래프 옵션(제목, 범례, 그래프 색 등)을 원하는 대로 정하기

### 제목
제목은 제 생일 7월 12일 최고, 최저 기온으로 설정했다. 한글 폰트는 '바탕'으로 설정했다.

### 범례 및 그래프 색
범례로 빨간색 선은 최고 기온, 초록색 선은 최저 기온을 나타내도록 했다. plt.legend(loc = 3)을 이용하여 왼쪽 아래에 범례가 위치하도록 했다.

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Batang')
plt.title('7월 12일 최고, 최저 기온')

plt.rcParams['axes.unicode_minus'] = False
plt.plot(high, color='r', label='최고 기온') # high 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.plot(low, color='g', label='최저 기온') # low 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.legend(loc=3)
plt.show()                    # 그래프 나타내기
```

## x축 이름, y축 이름

x축은 1904년부터 2022년까지의 7월 12일을 나타내는 의미로 '1904-07-12 ~ 2022-07-12(07-12)'라고 표현했고, y축은 기온을 의미하므로 '기온'으로 표시했다.

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Batang')
plt.title('7월 12일 최고, 최저 기온')
plt.xlabel('1904-07-12 ~ 2022-07-12(07-12)')
plt.ylabel('기온')

plt.rcParams['axes.unicode_minus'] = False
plt.plot(high, color='r', label='최고 기온') # high 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.plot(low, color='g', label='최저 기온') # low 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.legend(loc=3)
plt.show()                    # 그래프 나타내기
```
