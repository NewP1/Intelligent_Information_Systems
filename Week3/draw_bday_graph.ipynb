```python
import csv

f = open('incheon_weather.csv','r')
data = csv.reader(f)
next(data)
high = []                             # 최고 기온 데이터를 저장할 빈 리스트
low = []                              # 최저 기온 데이터를 저장할 빈 리스트

for row in data:                       
    if row[-1] != '' and row[-2] != '':                  # 최고 기온과 최저 기온 데이터 값이 존재한다면
        if int(row[0].split('-')[0] >= '1999'):           # 출생 년도 이후 데이터 비교
            if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':   # 9월 13일
                high.append(float(row[-1]))   # result 리스트에 최고 기온 값 추가
                low.append(float(row[-2]))
        
f.close()
# print(result)
# print(len(result))
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.plot(high, color='hotpink') # high 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.plot(low, color='skyblue') # low 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.show()                    # 그래프 나타내기
```
