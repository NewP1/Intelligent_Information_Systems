```python
from glob import glob

glob('지역*.xls')
```

```python
station_files = glob('지역*.xls')
station_files
```

```python
import pandas as pd

tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2) # 위에 두줄 없애기
    tmp_list.append(tmp)
tmp_list[0]
```

### tmp_list에 있는 모든 엑셀 파일의 내용들을 pd.concat을 이용하여 합친다.

```python
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations.info()
```

datafrmae의 형식은 아래와 같다.

```python
stations.head()
```

### 휘발유 가격을 담고 있는 dataframe과 경유 가겨을 담고 있는 dataframe을 각각 만들어 합치기로 했다.

gasoline_sel = 휘발유 dataframe

diesel_sel = 경유 dataframe 

```python
gasoline_sel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '휘발유']]
gasoline_sel
```

```python
diesel_sel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '경유']]
diesel_sel
```

휘발유, 경유 column의 이름을 '가격'으로 바꾸고 '종류' 열을 추가하여 휘발유인지 경유인지 적어준다.

```python
gasoline_sel.columns = ['주유소', '주소', '상표', '셀프여부','가격']
gasoline_sel['종류'] = '휘발유'
gasoline_sel
```

```python
diesel_sel.columns = ['주유소', '주소', '상표', '셀프여부','가격']
diesel_sel['종류'] = '경유'
diesel_sel
```

---

### gasoline_sel 과 diesel_sel 병합하기

```python
stations_sel = pd.concat([gasoline_sel, diesel_sel])
stations_sel
```

### '군/구' 열 추가하고 군/구 추출하여 데이터 넣기

```python
gu_list = []

for address in stations_sel['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)
stations_sel['군/구'] = gu_list
stations_sel
```

```python
# 군/구 목록 확인

stations_sel['군/구'].unique()
```

```python
stations_sel['상표'].unique()
```

## 예외 처리하기

```python
stations_sel.query('가격 == "-"')
```

```python
stations_sel = stations_sel.query('가격 != "-"').copy()
stations_sel
```

```python
# 가격을 숫자로 변경하기

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)
stations_sel['가격'] = price_list
stations_sel
```

## 그래프1) x축: 종류, y축: 가격, 색상: 셀프여부

### 그래프1) 결과 분석

전체적으로 경유는 대략 1900원, 휘발유는 대략 1600 ~ 1700원 대를 형성하고, 경유가 휘발유보다 높은 가격대를 형성하고 있다는 것을 볼 수 있다.

그래프 상으로는 셀프로 주유하지 않는 곳이 셀프가 되는 곳보다 가격이 조금 높은 것으로 보인다. 아무래도 사람을 쓰다보니 가격대가 좀 더 높은 것으로 생각된다. 그렇다고 해서 셀프가 아닌 곳이 셀프인 곳보다 절대적으로 가격이 높은 것은 아니다.

### 그래프3) 결과 분석

옹진군의 주유소들이 다른 군/구에 비해 가격이 훨씬 비싸다. 왜 그럴까?
이는 옹진군 주유소들은 섬에 위치해 있는 곳들이 있다. 주유소 수도 많지 않지만 적은 주유소들이 몇몇 섬들에 분포해 있기 때문에 기름값 가격대가 높아진 것 같다. 섬까지 유통하려면 아무래도 유통비가 더 들어 기름값이 높아지기 때문이다.

중구의 주유소는 인천 끝자락으로 영종도까지 영역이 잡혀있다. 가격대의 넓은 분포는 영종도 지역의 주유소에서의 기름 가격이 조금 높아 넓은 가격대가 형성된 것으로 보인다. 

다른 지역의 주유소들은 전체적으로 비슷한 가격대를 보인다.
