```python
from glob import glob


glob('지역*.xls')
```

```python
station_files = glob('지역*.xls')
station_files
```

```python
import pandas as pd

tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2) # 위에 두줄 없애기
    tmp_list.append(tmp)
tmp_list[0]
```

```python
stations = pd.concat(tmp_list)
```

```python
stations.shape
```

```python
stations.info()
```

```python
stations.head()
```

```python
stations_sel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '휘발유']]
stations_sel
```

```python
stations_sel.columns = ['주유소', '주소', '상표', '셀프여부','가격']
stations_sel
```

```python
gu_list = []

for address in stations_sel['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)
stations_sel['군/구'] = gu_list
stations_sel
```

```python
# 군/구 목록 확인

stations_sel['군/구'].unique()
```

## 예외 처리하기

```python
stations_sel.query('가격 == "-"')
```

```python
stations_sel = stations_sel.query('가격 != "-"').copy()
stations_sel
```

## 워닝이 뜬다.

### 워닝은 실행에 문제는 없음

아래의 이유는 기존에 있는 값에다가 뒤집어써서 문제가 있을 수 있다고 알려주는 내용

```python
# 가격을 숫자로 변경하기

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)
stations_sel['가격'] = price_list
stations_sel
```
