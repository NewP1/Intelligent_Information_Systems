# Selenium 이용하여 페이스북 로그인하기

```python
from selenium import webdriver

driver = webdriver.Chrome(r"C:\Users\jaehy\Downloads\driver\chromedriver.exe")
```

```python
main_url = "https://ko-kr.facebook.com/"
driver.get(main_url)
```

## 아이디는 facebookID

```python
id_xpath = '//*[@id="email"]'
driver.find_element('xpath', id_xpath).send_keys('facebookID')
```

## 비밀번호는 password123!

```python
pw_xpath = '//*[@id="pass"]'
driver.find_element('xpath', pw_xpath).send_keys('password123!')
```

## 로그인 버튼은 절대 경로로 넣어줬다. 상대경로는 안되던데..

```python
login_button = '/html/body/div[1]/div[1]/div[1]/div/div/div/div[2]/div/div[1]/form/div[2]/button'
driver.find_element('xpath', login_button).click()
```

```python
driver.close()
```
