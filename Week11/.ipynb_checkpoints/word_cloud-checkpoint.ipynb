```python
from konlpy.corpus import kolaw

fids = kolaw.fileids()
fobj = kolaw.open(fids[0])

print(fobj.read(140))
ko_doc = fobj.read() # 전문 가져오기
```

```python
from wordcloud import WordCloud

wc = WordCloud(background_color='white', max_words=2000)
wc = wc.generate(ko_doc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

왜 네모로 나올가?

font_path를 이용하자

```python
wc = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\malgun.ttf')
wc = wc.generate(ko_doc)

plt.figure(figsize=(12,8))
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

interpolation 보간법!!!!!!!!!!!!!!

```python
from collections import Counter

wc = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\malgun.ttf')
count = Counter(ko_doc.split())
wc = wc.generate_from_frequencies(count)
count.most_common(20)
```

```python

plt.figure(figsize=(12,8))
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

# 워드 클라우드 그리기(2)

## 의미없는 단어 처리하기

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
ko_doc_noun = okt.nouns(ko_doc)
print(ko_doc_noun)
```

```python
count_noun = Counter(ko_doc_noun)

wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\malgun.ttf')
wc_noun = wc_noun.generate_from_frequencies(count_noun)
count_noun.most_common(20)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc_noun, interpolation='bilinear')
plt.show()
```

# stopword list 만들기

```python
count_noun.most_common()
```

```python
count_noun.pop('제')
```

```python
count_noun.most_common()
```

```python
stopword_list = []

for noun_word in count_noun:
    if len(noun_word) == 1:
        stopword_list.append(noun_word)

for stopword in stopword_list:
    count_noun.pop(stopword)
```

```python
count_noun.most_common()
```

```python
wc_noun = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\malgun.ttf')
wc_noun = wc_noun.generate_from_frequencies(count_noun)

plt.figure(figsize=(12,8))
plt.imshow(wc_noun, interpolation='bilinear')
plt.show()
```

```python
import numpy as np
from PIL import Image

img_mask = np.array(Image.open('law.jpg'))
wc_img = WordCloud(background_color='white', max_words=2000, font_path=r'C:\Windows\Fonts\malgun.ttf', mask=img_mask)
wc_img = wc_img.generate_from_frequencies(count_noun)

plt.figure(figsize=(12,8))
plt.imshow(wc_img, interpolation='bilinear')
plt.show()
```
